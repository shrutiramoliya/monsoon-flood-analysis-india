# Spatial Analysis of Urban Flood Events in India (2010–2025)

## Geographic Visualization and Flood Hotspot Analysis

This notebook explores the spatial distribution of urban flood events across Indian cities and states using geographical coordinates and mapping techniques.

## Objectives

The objectives of this analysis are:

- Visualize geographic distribution of flood events
- Identify flood-prone urban regions
- Explore spatial concentration patterns
- Create flood hotspot visualizations across India

```python
import pandas as pd
import folium
```

## Loading the Dataset

```python
df = pd.read_excel("../data/India_Flood_Dataset_Final.xlsx")
```

```python
df.head()
```

## Urban Flood Event Mapping

```python
india_map = folium.Map(location=[22.5, 80.9], zoom_start=5)

india_map
```

```python
for index, row in df.iterrows():
    
    folium.CircleMarker(
        location=[row["LATITUDE"], row["LONGITUDE"]],
        radius=5,
        popup=f"{row['CITY']} ({row['YEAR']})",
        color="blue",
        fill=True,
        fill_color="blue"
    ).add_to(india_map)

india_map
```

### Observation

Flood events appear concentrated across major urban and coastal regions, indicating spatial clustering of urban flood vulnerability in India.

```python
from folium.plugins import HeatMap
```

```python
heat_data = df[["LATITUDE", "LONGITUDE"]].values.tolist()
```

```python
heatmap = folium.Map(location=[22.5, 80.9], zoom_start=5)

HeatMap(heat_data).add_to(heatmap)

heatmap
```

### Observation

The heatmap highlights spatial concentration of urban flood events across specific regions of India. Major urban and coastal areas appear as dominant flood hotspots.

```python
df_2020 = df[df["YEAR"] == 2020]
```

```python
map_2020 = folium.Map(location=[22.5, 80.9], zoom_start=5)

for index, row in df_2020.iterrows():

    folium.CircleMarker(
        location=[row["LATITUDE"], row["LONGITUDE"]],
        radius=6,
        popup=f"{row['CITY']} ({row['YEAR']})",
        color="red",
        fill=True,
        fill_color="red"
    ).add_to(map_2020)

map_2020
```

### Observation

Spatial flood distribution varies across years, indicating changing regional concentration and urban flood exposure patterns over time.
